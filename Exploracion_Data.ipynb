```python
# Librerias necesarias
import os
import pandas as pd
```

### Carga de datos

```python
# Cargar en una lista los nombres de todos los archivos
fileList = os.listdir()
 
print(fileList)
```

```python
listCsv = [file for file in fileList if file[0:9] == "PlazaVea_"]
print(listCsv)
```

```python
# Crear DataFrame vacio
df = pd.DataFrame()
 
# Recorrer la lista de archivos y unirlos
for file in listCsv:
    ftemp = pd.read_excel(file)
    df = pd.concat([df, ftemp], axis=0)

df.head(5)
```

### Tratamiento de datos

```python
df['Fecha'] = df['Fecha'].str[:10]
df
```

```python
df['Fecha'] = pd.to_datetime(df['Fecha'])
df
```

```python
# Obtener un codigo unico por producto
df['Codigo_Producto'] = pd.factorize(df['Nombre_Producto'])[0] + 1
```

```python
# Ordenar el DataFrame por 'Fecha' y 'Precio_Real' en orden ascendente
df_origen = df.sort_values(['Fecha', 'Precio_Real'],ascending=[True, True])

# Para que solo exista un unico resgistro por fecha y producto
df_origen['RowNumber'] = df_origen.groupby(['Fecha','Nombre_Producto']).cumcount() + 1

df_origen = df_origen[df_origen['RowNumber'] == 1]
df_origen = df_origen.drop('RowNumber', axis=1)
```

### Análisis de datos

##### Ejemplo de manipulación de datos

```python
import pandas as pd

# DataFrame
df_example = pd.DataFrame({'Grupo': ['A', 'A', 'B', 'B', 'B', 'C', 'C'],
                   'Valor': [10, 20, 30, 40, 50, 70, 60]})

# Ordenar el DataFrame por 'Grupo' y 'Valor' en orden ascendente
df_sorted = df_example.sort_values(['Grupo', 'Valor'],ascending=[True, False])

# Obtener los códigos numéricos únicos para cada letra en el campo "Grupo"
df_sorted['Codigo'] = pd.factorize(df_sorted['Grupo'])[0] + 1

# Aplicar la operación OVER de suma acumulativa
df_sorted['RowNumber'] = df_sorted.groupby('Grupo').cumcount() + 1
df_sorted['SumaAcumulativa'] = df_sorted.groupby('Grupo')['Valor'].cumsum()
df_sorted['ValorAcumulado_unit'] = df_sorted.groupby('Grupo')['Valor'].transform('count')
df_sorted['ValorAcumulado_sum'] = df_sorted.groupby('Grupo')['Valor'].transform('sum')
df_sorted['ValorAcumulado_mean'] = df_sorted.groupby('Grupo')['Valor'].transform('mean')

print(df_sorted)
```

##### Análisis de precios

```python
df_analysis = df_origen[df_origen['Fecha'] != df_origen['Fecha'].max()]
print(df_analysis['Fecha'].max())
```

```python
fecha_corte = df_origen['Fecha'].max()
```

```python
# Nueva logica de resumen

df_resumen = df_analysis.groupby(['Codigo_Producto','Nombre_Producto']).agg({'Precio_Real': ['mean','min','max']})
df_resumen.columns = ['Media_Precio', 'Minimo_Precio', 'Maximo_Precio']
df_resumen.head()
```

```python
porcentaje_oferta = 0.20
df_resumen['var_superior'] = df_resumen['Media_Precio']*(1 + porcentaje_oferta)
df_resumen['var_inferior'] = df_resumen['Media_Precio']*(1 - porcentaje_oferta)
df_resumen.head()
```

```python
df_precio_altual = df_origen[df_origen['Fecha'] == df_origen['Fecha'].max()]

df_join = df_resumen.merge(df_precio_altual, how='outer', on='Codigo_Producto')
```

```python
df_join.columns
```

```python
df_join = df_join[['Codigo_Producto','Categoria','Subcategoria','Nombre_Producto','Precio_Real','var_superior', 'var_inferior', 'Precio_Final', 'Media_Precio', 'Minimo_Precio', 'Maximo_Precio']]
df_join.head(3)
```

```python
def diferencia_extremo(df, precio, superior, inferior):
    if df[precio] <= df[inferior]:
        return 'Alerta de precio'
    else:
        if df[inferior] <= df[precio] <= df[superior]:
            return 'Sin cambios'
        else:
            return 'Alza de precio'

df_join['Criterio'] = df_join.apply(diferencia_extremo, axis=1, precio='Precio_Real',superior='var_superior',inferior='var_inferior')
```

```python
df_join[df_join['Criterio'] == 'Alerta de precio'].sort_values(['Precio_Real'],ascending=[False])
```
